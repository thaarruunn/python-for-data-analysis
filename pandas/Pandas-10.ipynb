working with dates and times , ie : Time Series

dates like "2026-06-22" look like dates to your human eyes, but Pandas just sees them as generic text (strings or object types).

to fix it , Your very first step with any date column should always be converting it into a true Pandas Datetime object:
# Convert text to actual datetime data
df['Date'] = pd.to_datetime(df['Date'])

Once your column is a true datetime object, Pandas gives you a magical helper called .dt. This lets you extract practically any component of the date instantly without complex string slicing.

```python
df['Year']  = df['Date'].dt.year          # 2026
df['Month'] = df['Date'].dt.month         # 6
df['Day']   = df['Date'].dt.day_name()    # 'Monday'
df['Hour']  = df['Date'].dt.hour          # 19 (7 PM)
```

slicing with time:

```python
df = df.set_index('Date').sort_index()

# Grab all data from the entire year of 2025
print(df.loc['2025'])

# Grab everything from June 2025 to March 2026
print(df.loc['2025-06':'2026-03'])
```

For advanced time-series functions to work efficiently, the date column must become the index, and it must be sorted.

```python
#hence ,
df = df.set_index('Timestamp').sort_index()
```

```python
# Forward fill: Carries the last known value forward into the missing slot
df['Sales'] = df['Sales'].ffill()

# Backward fill: Replaces missing value with the next available value
df['Sales'] = df['Sales'].bfill()
```

What Timedelta Does
While a Datetime represents a specific pinpoint moment in time (e.g., June 22, 2026, at 7:45 PM), a Timedelta represents a duration or span of time (e.g., 3 days, 4 hours, and 12 minutes).

You create a Timedelta automatically whenever you subtract one date column from another.

```python
eg1:
# Subtracting two Datetimes results in a Timedelta column
df['Delivery_Duration'] = df['Delivery_Date'] - df['Order_Date']

print(df['Delivery_Duration'].head())
# Output: 
# 0   3 days 04:00:00
# 1   1 days 12:30:00
```

```python
eg2. 
# Find out what date it will be exactly 14 days after the order
df['Deadline'] = df['Order_Date'] + pd.Timedelta(days=14)
```
