Matplotlib

 passing summary.index and summary.values tells the plotting function exactly where to place the data points (the categories/labels)
 
  and how high or long those points should be (the actual numbers).
  
  here is exactly what each one gives to Matplotlib:
  
  1. summary.index $\rightarrow$ The Labels or Coordinates (usually the X-axis)
  
  The .index contains the identifiers of your rows. In Matplotlib, these are typically used as your categorical labels or your independent variables.
  
  In a Bar Chart: These become the labels written under each bar on the x-axis (e.g., 'Platform A', 'Platform B').
  
  In a Line/Scatter Plot: These provide the horizontal coordinates ($x$-values) for where each point is plotted.
  
  2. summary.values $\rightarrow$ The Data Points or Magnitudes (usually the Y-axis)The .values is a raw NumPy array containing the actual numeric data. 
  
  Matplotlib uses this to determine the scale, height, or position of the visual elements.
  
  In a Bar Chart: This determines the physical height of each bar.
  
  In a Line/Scatter Plot: This provides the vertical coordinates ($y$-values) for the points.
