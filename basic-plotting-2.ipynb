<strong>Creating matplotlib charts</strong>

```python
import matplotlib.pyplot as plt
```

```python
#data
platforms = [
    "Blinkit",
    "Zepto",
    "Instamart",
    "BB Now"
]

orders = [
    12000,
    15000,
    9000,
    7000
]
```

*sample question:*

*Question:*

*Which platform gets most orders?*

--*Category comparison*

*Chart?*

*Bar chart*

```python
#code to solve the above problem :

fig,axis=plt.subplots(figsize=(6,5)) #(width , height)
axis.bar(platforms,orders)
axis.set_title("Comparision of most orders")
axis.set_xlabel("sales amount")
axis.set_ylabel("Companies")
plt.show() #very necessary to show visualization!
```

<strong>Different Kinds of Charts</strong>

Line chart : used if theres change over continuous order/time

```python
#example data :
months=[
"Jan","Feb","Mar","Apr"
]

sales=[
100,150,130,200
]
```

```python
fig,axis=plt.subplots(figsize=(6,4))
axis.plot(months,sales,marker='o')
axis.set_title("Sales over period of time")
axis.set_xlabel("months")
axis.set_ylabel("sales")
plt.show()
```

marker="o"

Shows points:

Jan ●────● Feb

Useful because viewers see exact observations.

<strong>Scatter Plot</strong>

*example q:*

*does discount affect order value?*

```python
#data
discount=[
5,10,20,30
]

order_value=[
200,350,700,900
]
```

```python
fig, ax = plt.subplots(figsize=(8,5))

ax.scatter(
    discount,
    order_value
)

ax.set_xlabel(
    "Discount %"
)

ax.set_ylabel(
    "Order Value"
)

ax.set_title(
    "Discount vs Order Value"
)

plt.show()
```

<strong>Histogram</strong>

*example q*:
*How are delivery times distributed?*

```python
#data
times=[
10,12,13,15,20,30,40
]
```

```python
fig,axis=plt.subplots() #use figsize if wanted :)
axis.hist(times,bins=5)
plt.show()
```

bins

Histogram groups values.

Example:

0-10 min

10-20 min

20-30 min

Those groups are bins.

Too few bins:

lose detail.

Too many:

messy.
