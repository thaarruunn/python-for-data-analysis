Merges and Joins in Pandas

```python
#1 Join :

pd.merge(students,marks,on='id')
```

```python
#2 Inner Join:

pd.merge(students,marks,on='id',how='inner')

#3 Left Join:

pd.merge(students,marks,on='id',how='left')

#4 Outer Join:

pd.merge(students,marks,on='id',how='outer')
```
