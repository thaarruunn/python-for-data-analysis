```python
Pandas has TWO core data structures. 
    (1) Series
    (2) DataFrame
```

```python
import pandas as pd
```

```python
# Example for Series in Pandas
data=pd.Series([1,2,3],index=[0,1,2])
print(data)
```

```python
wtv is on the left side is called as index. right side are the values.
```

```python
marks=[95,99,83,80,96]
df=pd.Series(marks,index=['chem','cs','maths','phy','eng'])
print(df)
```

```python
Dataframe is rows + columns + index. its like an excel sheet or a SQL table or 
a CSV file
```

```python
Data={"Name":['Messi','Dube','Amex'],"SPM":[1,0.55,0.35]}
people=pd.DataFrame(Data)
print(people)
```
