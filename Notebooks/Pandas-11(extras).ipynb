(1)Saving & Handling Missing Data:
Saving Files (.to_csv() vs .to_excel())
index=False: Crucial for almost every export. It prevents pandas from saving row numbers (0, 1, 2...) as an annoying, unnamed first column in your Excel/CSV file.

na_rep='N/A': Stands for "Null Representation." It replaces empty/blank spaces (NaN) in your data with a clean string like 'N/A' or 'NULL'.

```python
# Save as a clean Excel file with a custom sheet name
df.to_excel('sales_report.xlsx', index=False, sheet_name='Q1_Sales')

# Save as a clean CSV file with explicit missing value text
df.to_csv('sales_report.csv', index=False, na_rep='N/A')
```

(2)Modifying DataFrames (The Golden Rules)
Why things don't delete/change
Pandas protects your data by default. Methods like .drop(), .sort_values(), and .fillna() only return a copy of the modified data; they do not change the original DataFrame unless you explicitly tell them to.

```python
# ❌ INCORRECT: This will run, but print(df) will show the column is still there!
df.drop('rank square', axis=1) 

#  CORRECT: Reassign it back to your variable to save it
df = df.drop('rank square', axis=1)
```

(3)Selecting & Updating Data (.loc vs .iloc)
.loc[row_label, col_label]: Label-based. Uses actual names. Slicing like 'A':'C' includes C.

.iloc[row_pos, col_pos]: Integer-based. Uses numeric positions (0, 1, 2...). Slicing like 0:3 excludes position 3 (only grabs 0, 1, and 2).

```python
#  The best way to update a single cell safely
df.loc[0, 'country'] = 'Canada'

# The best way to update based on a condition
df.loc[df['age'] > 60, 'status'] = 'Retired'
```

Summary:

1. Data Export (.to_csv, .to_excel)
Syntax: df.to_csv('filename.csv', index=False, na_rep='N/A')

Uses: Saves your cleaned DataFrame to an external file. index=False prevents row numbers from creating an unwanted column, and na_rep replaces blanks with clean text like 'N/A'.

2. Modifying Rows & Columns (.drop, .sort_values)
Syntax: df = df.drop(['Col1', 'Col2'], axis=1) OR df = df.drop(columns=['Col1'])

Uses: Permanently deletes columns from your DataFrame (axis=1 points to columns).

Syntax: df = df.sort_values(by='ColumnName', ascending=False)

Uses: Sorts your dataset based on a specific column in descending order.

3. Updating Values (.loc, np.where)
Syntax: df.loc[df['Age'] > 60, 'Status'] = 'Retired'

Uses: The most reliable way to update specific rows based on an "If" condition.

Syntax: df['Status'] = np.where(df['Score'] >= 50, 'Pass', 'Fail')

Uses: Best for quick "If-Else" column creation across the entire dataset.

4. Index Management (.reset_index)
Syntax: df = df.reset_index(drop=True)

Uses: Fixes row numbers (0, 1, 2...) after they get scrambled from sorting or dropping rows. drop=True deletes the old scrambled numbers.

5. Frequency & SQL Equivalents (.value_counts, .groupby)
Syntax: df[['Col1', 'Col2']].value_counts().reset_index()

Uses: The fastest way to replicate SQL's SELECT Col1, Col2, count(*) ... GROUP BY Col1, Col2.

Syntax: df.groupby(['Col1', 'Col2']).size().reset_index(name='count')

Uses: The literal translation of SQL GROUP BY when you want to compute size counts.

6. Utility Shortcuts (.isin, .unique, .rename)
Syntax: df[df['Country'].isin(['USA', 'UK'])]

Uses: Replicates SQL's WHERE Country IN ('USA', 'UK'). Saves you from writing multiple | (OR) operators.

Syntax: df['Column'].unique() (returns an array) / df['Column'].nunique() (returns a count)

Uses: Instantly checks distinct categories or counts how many unique values exist in a column.

Syntax: df = df.rename(columns={'old_name': 'New_Name'})

Uses: Cleans up or fixes typos in specific column names using a dictionary.
