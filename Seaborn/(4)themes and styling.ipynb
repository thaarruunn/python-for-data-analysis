seaborn themes

*The set_theme() Function*

*This is usually the first Seaborn line in a notebook.*

*Once you run it,*

*every Seaborn plot (and many Matplotlib plots) follows that style until you change it.*

*Think of it as setting the visual environment for your notebook.*

---

DARK GRID

```python
sns.set_theme(style="darkgrid")

#gray background
#grid lines visible
#easy to compare values
```

---

WHITE GRID

```python
sns.set_theme(style="whitegrid")

#white background
#clean
```

---

White

```python
sns.set_theme(style="white")

#white background
#no grid
#minimalistic
```

---
Dark

```python
sns.set_theme(style="dark")

#dark gray background
#no grid
```

---

Ticks

```python
sns.set_theme(style="ticks")

#white background
#small tick marks
#minimal grid
```

*color palettes*

*a color is one color.*

*a palette is a coordinated collection of colors.*

*instead of randomly choosing colors, Seaborn provides palettes designed to work well together.*

```python
sns.set_palette("deep")
```

*other common palettes include:*

"deep" (balanced, default)

"muted" (softer colors)

"bright" (high contrast)

"pastel" (lighter tones)

"dark" (darker shades)

"colorblind" (optimized for color vision deficiencies)
