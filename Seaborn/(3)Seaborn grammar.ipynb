<strong>the universal seaborn pattern:</strong>

```python
sns.someplot(
    data=df,
    x="column1",
    y="column2",
    hue="column3"
)

#Think of hue as:

"Split each category using another variable."
```

whether you're creating a:

Bar Plot

Scatter Plot

Box Plot

Line Plot

Violin Plot

Strip Plot

the structure stays almost identical.

only someplot changes.

*eg:1*

    y="Rating"
)

*eg:2*

Does delivery time affect rating?

sns.scatterplot(
    data=df,
    x="Delivery_Time",
    y="Rating"
)


────────────────────────────────────────────

1. data=

Specifies the DataFrame containing the data.

Example:
data=df

Seaborn automatically searches this DataFrame for the specified columns.

────────────────────────────────────────────

2. x=

Specifies the column placed on the x-axis.

Example:
x="Platform"

Only the column name is provided because Seaborn already knows which DataFrame to use.

────────────────────────────────────────────

3. y=

Specifies the column placed on the y-axis.

Example:
y="Rating"

────────────────────────────────────────────

4. hue=

Adds another variable by splitting the data into groups, usually represented using different colours.

Example:
hue="Premium"

This allows one visualization to display an additional categorical variable.
