Seaborn mental model

<strong>step 1 : ask the question

*For example:*

*Which platform has the highest average rating?*

*or*

*Does delivery time affect customer ratings?*

*or*

*Are order values normally distributed?*

*Notice that these are questions, not plots.*

<strong> step 2: identify the variables

*once you have a question , identify the variables involved.*

*eg ; whcih platform has the highest rating?*

*variables ; platform , rating*

*everything else in the DataFrame is irrelevant for this graph*

<strong>step 3: identify the variable types

*every column belongs to a type.*

*There are two main types:*

<Strong>(1) categorical variables:</strong>

*these represent labels or groups*

*eg :*

*Platform*

*City*

*Gender*

*Delivery Speed Category*

*Payment Method*

<strong>you cant calculate the average of blinkit , hence categories answer : which group?

(2)<strong>Numerical variables:

*they represent measurable quantities*

*examples:*

*delivery time*

*order value*

*distance*

*avg rating*

*with these, u can calculate mean , median , std dev etc*

<strong>why this matters?

*Because Seaborn chooses plots based on variable types, not just the question.*

*eg:*

<strong>one categorical + one numerical</strong>

*platform --> average rating* 

*possible plots: bar - box - violin - strip*


<strong>two numerical variables</strong>

*delivery time -->rating*

*possible plots: scatterplot - regression - lineplot*

<strong>one numerical variable</strong>

*delivery time --> histogram ; etc*
