<strong>Matplotlib Customization & Professional Visualization</strong>

The goal:

Not making graphs "pretty."

The goal:

Make graphs easier and faster to understand.

*(1) Figure Size (figsize)*

```python
fig, ax = plt.subplots(figsize=(10,5))

(width,height)
```

(2)Title :

#should be a GOOD title name. :



```python
ax.set_title("Total Revenue by Quick Commerce Platform",fontsize=14)
```

(3)Axis labels

```python
ax.set_xlabel("Platform")
ax.set_ylabel("Revenue (₹)")
```

(4)Tick Labels

*Ticks are the numbers/categories on axes, to ensure no overlapping happens*

```python
plt.xticks(rotation=45)

#leads to better readability
```

(5) Grid Lines



```python
axis.grid(True)
```

(6)Legends

*used when two or more things exists , and to know what is what*

```python
ax.plot(
    months,
    blinkit_sales,
    label="Blinkit"
)

ax.plot(
    months,
    zepto_sales,
    label="Zepto"
)


ax.legend() #very important
```

(7)Line Chart customization

```python
ax.plot(
    x,
    y,
    marker="o", # data points
    linewidth=2,# thickness
    linestyle="--" #style
)
```

(8)Line plot customization

s=Size.
alpha=Transparency.

Important when many points overlap.

Example:10000 customerss,without transparency:black blob. with transparency:density visible.

```python
ax.scatter(
    x,
    y,
    s=50,
    alpha=0.7
)
```

(9) Tight layout: use before plt.show() to ensure ease of readability

```python
plt.tight_layout()
```

(10) colours :

```python
colors = ["blue","orange","green","red"]
ax.bar(platforms,revenue,color=colors)

#use if multiple colors required else

ax.bar(platforms,revenue,color='red')


```

xticks vs tight layout

| Problem                   | Solution              |
| ------------------------- | --------------------- |
| Labels overlap each other | `tick_params(axis='x',rotation=45)` |
| Labels/title get cut off  | `tight_layout()`      |
| Subplots too close        | `tight_layout()`      |
| Want custom tick names    | `xticks()`            |
