*Subplots and Analytical Dashboards*

```python
import matplotlib.pyplot as plt
```

GOAL :

(1) *Creating multiple axes*

```python
fig, ax = plt.subplots(2,2)
```

*the above line of code means 2 rows and 2 columns = 4 axes*

ax[0,0]    ax[0,1]


ax[1,0]    ax[1,1]

*with each axes having its own graph*

```python
platforms= [
    "Blinkit",
    "Zepto",
    "Instamart",
    "BBNow"
]

revenue= [
    500,
    650,
    450,
    300
]

delivery_time= [
    10,12,15,18,25,30,40
]

discount= [
    5,10,15,20,25
]

order_value= [
    200,300,450,600,800
]
```

```python
fig,ax=plt.subplots(2,2,figsize=(12,8))
```

*creating a revenue dashboard on top left*

```python
ax[0,0].bar(
    platforms,
    revenue
)

ax[0,0].set_title(
    "Revenue by Platform"
)
```

*creating a delivery distribution dashboard on top right*

```python
ax[0,1].hist(
    delivery_time,
    bins=10
)

ax[0,1].set_title(
    "Delivery Time Distribution"
)
```

*creating a relationship graph*

```python
ax[1,0].scatter(
    discount,
    order_value
)

ax[1,0].set_title(
    "Discount vs Order Value"
)
```

*boxplot*

```python
ax[1,1].boxplot(
    delivery_time
)

ax[1,1].set_title(
    "Delivery Outliers"
)
```

```python
plt.tight_layout()
plt.show()
```

*plt.tight_layout() ensures no overlapping*

```python

fig, ax = plt.subplots(2, 2, figsize=(10, 8))

ax[0,0].bar(platforms, revenue)
ax[0,0].set_title("Revenue by Platform")

ax[0,1].hist(delivery_time, bins=10)
ax[0,1].set_title("Delivery Time Distribution")

ax[1,0].scatter(discount, order_value)
ax[1,0].set_title("Discount vs Order Value")

ax[1,1].boxplot(delivery_time)
ax[1,1].set_title("Delivery Outliers")

plt.tight_layout()
plt.show()
```

USING SHAREY - same axis

*The main need for sharey=True (or sharey in general) is to ensure accurate visual comparison across different subplots by locking them to the exact same vertical scale*

*By default, Matplotlib looks at each plot individually and auto-scales its y-axis to fit only that specific dataset. This can easily lead to misinterpreting your data*

*Here is why it is incredibly useful:*

*1. It Prevents Visual Deception*
*If you are comparing two bar charts or line graphs side-by-side:*

*Without sharey: Plot A might have a y-axis ranging from 0 to 10, and Plot B might range from 0 to 1,000. Visually, a bar of height "8" on Plot A will look just as tall as a bar of height "800" on Plot B.*

```python
fig, ax = plt.subplots(
    2,
    2,
    sharey=True
)
```

*all share same y-axis*
