*different phases of a data scientist-analyst's workflow*

<strong>phase 1 : understand the problem</strong>

*Never open the CSV first.*

*Ask:*

*What is the company?*

*What is the business problem?*

*What decisions are they trying to make?*


*Example:*

*Quick Commerce*

*Business Problem:*

*Why are customer ratings decreasing?*

*That's your objective.*

*Everything else serves this.*

<strong>phase 2: understand the dataset</strong>

*before writing the code , ask*

*what are the rows , columns , missing values etc*

*use:*

```python
df.head()

df.info()

df.describe()

df.isnull().sum()

df.nunique()
```

<strong>phase3: Data Cleaning</strong>

*typical tasks include: *

*missing null values*

*duplicate values*

*wrong spellings*

*extra spaces*

<strong>phase4: Univariate Analysis</strong>

*Univariate Analysis refers to analyzing ONE variable*

*questions include:* 

*average?*

*median?*

*min , max?*

*of ONE variable , lets say order value*

<strong>charts often used include histogram - box plot</strong>

<strong>phase5: Bivariate Analysis</strong>

*Analyzes TWO variables*

*eg :*

*discount v/s order value*

*delivery fee v/s rating*

*platform v/s revenue*

<strong>charts often include bar - scatter - boxplot</strong>

<strong>phase6: Multivariate Analysis</strong>

*includes analysis of 3 or more variables*

Example:

Platform

↓

City

↓

Average Delivery Time

↓

Customer Rating

<strong>phase7: Business questions</strong>

*dont ask* :

*what is the average order value*

*ask:*

*what platform generates the highest revenue*

*dont ask:*

*average delivery time*

*instead*

*which platform should increase logistics operations to reduce delivery time*

<strong>phase8: Visualization</strong>

*choose chart(s) based on questions.*

<strong>phase9: Insights</strong>

*every graph should produce an insight*

*eg:*

*observation:*

*blinkit average delivery: 14 minutes*

*insight:*

*blinkit's logistics and delivery operations are more efficient than other competitors*

<strong>phase10: Recommendations</strong>

*insights without recommendations are incomplete*

*example:*

*delivery delays increase sharply after 8pm*

*Recommendation*:

*increase no:of riders after 8pm / evening time due to high demand*

<strong>example qn:</strong>

*does coupon usage increase spending*
