Quartiles : Split data into four parts.

Q1: 25% of data below.
Q2: Median. 50%.
Q3:75%. Example: Q1 = 10
Median = 20
Q3 = 40

Means:

50% of data lies between:

10 and 40.

(2)IQR ⭐

Very important.

Formula:

IQR = Q3 - Q1

Measures spread.

Outliers

Usually:

Lower limit:

Q1 - 1.5*IQR

Upper limit:

Q3 + 1.5*IQR

Anything outside:

potential outlier.
